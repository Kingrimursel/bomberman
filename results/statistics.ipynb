```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
better_bomb = np.load('data/action_BETTER_BOMB_POSSIBLE.npy')
good_bomb  = np.load('data/action_GOOD_BOMB.npy')
missed_bomb = np.load('data/action_BOMB_DROPPED.npy')
```

```python
# performance plot
fig, ax = plt.subplots(1, 3, figsize=(15, 5), sharey=True)

x = np.arange(len(better_bomb))
ax[0].bar(x, better_bomb, width=1, align='center', color='m', label='Frequency of Event')
ax[1].bar(x, good_bomb, width=1, align='center', color='navy', label='Frequency of Event')
ax[2].bar(x, missed_bomb, width=1, align='center', color='cornflowerblue', label='Frequency of Event')

ax[0].set_ylabel('Frequency of Events', fontsize=14, fontname='Helvetica')

ax[0].set_xlabel('Game Round', fontsize=14,  fontname='Helvetica')
ax[1].set_xlabel('Game Round', fontsize=14, fontname='Helvetica')
ax[2].set_xlabel('Game Round', fontsize=14, fontname='Helvetica')

ax[0].set_xlim(0, 50)
ax[1].set_xlim(0, 50)
ax[2].set_xlim(0, 50)

ax[0].set_title('Frequency of BETTER_BOMB', fontsize=18, fontname='Helvetica')
ax[1].set_title('Frequency of GOOD_BOMB',  fontsize=18, fontname='Helvetica')
ax[2].set_title('Frequency of BOMB_DROPPED', fontsize=18, fontname='Helvetica')

fig.tight_layout()

plt.savefig('images/bomb_analysis.png', dpi=600)
plt.show()
```
