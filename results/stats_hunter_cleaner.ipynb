```python
import numpy as np
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
fname = 'cleaner_hunter_comp'
with open('data/' + fname + '.json', 'r') as file:
    data = json.load(file)
```

```python
names  = []

suicides = []
score    = []
coins    = []
kills    = []
times    = []

for agent in data['by_agent']:
    try: 
        suicides.append(data['by_agent'][agent]['suicides'])
    except:
        suicides.append(0)
    try:
        score.append(data['by_agent'][agent]['score'])
    except:
        score.append(0)
    try:
        coins.append(data['by_agent'][agent]['coins'])
    except:
        coins.append(0)
    try:
        kills.append(data['by_agent'][agent]['kills'])
    except:
        kills.append(0)
    try:
        times.append(data['by_agent'][agent]['time'])
    except:
        times.append(0)
        
    strings = ['_', '0', '1', '2', '3']
    replace = [' ', '', '', '', '']
    for s, r in zip(strings, replace):
        agent = agent.replace(s, r)
    names.append(agent)
        
suicides = np.array(suicides)[:3]
score    = np.array(score)[:3]
coins    = np.array(coins)[:3]
kills    = np.array(kills)[:3]
times    = np.array(times)[:3]
```

```python
agents     = np.array([score, coins, 5*kills, 5*suicides, times]).T
categories = np.array(['score', 'coin score', 'kill score', r'5 $\cdot$ suicides', 'time'])

# plotting
plt.rc('font', size=20)
plt.rc('font', family='Helvetica')
plt.figure(figsize=(15, 9))

x = np.arange(len(categories))
plt.bar(x-0.225, agents[0], width=0.2, color='m', label='own hunter')
plt.bar(x, agents[1], width=0.2, color='navy', label='own cleaner')
plt.bar(x+0.225, agents[2], width=0.2, color='cornflowerblue', label='rule based agent')

plt.xticks(x, categories)
plt.ylabel('counts')
plt.legend()

#plt.savefig('images/hunter_against_cleaner_score.png', dpi=500)
plt.show()
```
